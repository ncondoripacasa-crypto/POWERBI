# Análisis de Herramientas BI
Este cuaderno es el insumo principal del proyecto. Descarga los datos de Kaggle y realiza una exploración inicial.

```python
import kagglehub
import pandas as pd
import os

# Descargar la última versión del dataset
path = kagglehub.dataset_download("karlitaestre/herramientas-bi")
print("Ruta de los archivos del dataset:", path)
```

```python
# Listar archivos CSV descargados
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
print("Archivos CSV encontrados:", csv_files)

if csv_files:
    # Cargar el primer archivo CSV encontrado
    df = pd.read_csv(os.path.join(path, csv_files[0]))
    display(df.head())
else:
    print("No se encontraron archivos CSV en el dataset.")
```
